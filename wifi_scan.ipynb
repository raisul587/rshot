```python
import time
import pywifi
from pywifi import const

def scan_wifi_networks():
    # Initialize the wifi interface
    wifi = pywifi.PyWiFi()
    ifaces = wifi.interfaces()
    if not ifaces:
        print("No Wi-Fi interface found.")
        return []
    iface = ifaces[0]
    
    # Start a scan
    iface.scan()
    # Wait a few seconds for the scan to complete
    time.sleep(5)
    
    # Retrieve scan results
    results = iface.scan_results()
    networks = []
    
    for network in results:
        networks.append({
            "ssid": network.ssid,
            "bssid": network.bssid,
            "signal": network.signal,   # signal strength in dBm
            "frequency": network.freq
        })
    
    return networks

def display_networks(networks):
    if not networks:
        print("No networks found.")
        return
    
    for idx, net in enumerate(networks, start=1):
        print(f"Network {idx}:")
        print(f"  SSID      : {net['ssid']}")
        print(f"  BSSID     : {net['bssid']}")
        print(f"  Signal    : {net['signal']} dBm")
        print(f"  Frequency : {net['frequency']} MHz")
        print("-" * 40)

def main():
    networks = scan_wifi_networks()
    display_networks(networks)

if __name__ == "__main__":
    main()
```
